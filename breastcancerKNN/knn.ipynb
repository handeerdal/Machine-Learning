```python
X = df.drop('diagnosis', axis=1)  # 'diagnosis' sütunu haricindeki tüm sütunları X olarak tanımla feature
y = df['diagnosis']  # 'diagnosis' sütununu y olarak tanımla label
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3) # n_neighbors = k
knn.fit(X_train,y_train)
prediction = knn.predict(X_test)
print("K=3 için Test verilerimizin doğrulama testi sonucu ", knn.score(X_test, y_test))
```

```python
# k kaç olmalı ?
# en iyi k değerini belirleyelim..
sayac = 1
for k in range(1,11):
    knn_yeni = KNeighborsClassifier(n_neighbors = k)
    knn_yeni.fit(X_train,y_train)
    print(sayac, "  ", "Doğruluk oranı: %", knn_yeni.score(X_test,y_test)*100)
    sayac += 1
    
    
```

```python
from sklearn.metrics import confusion_matrix
# Modelin tahminlerini kullanarak karmaşıklık matrisini hesapla
conf_matrix = confusion_matrix(y_test, prediction)
print("Confusion Matrix:\n", conf_matrix)
```

```python
from sklearn.metrics import precision_score, recall_score

# Precision ve recall değerlerini hesapla
precision = precision_score(y_test, prediction)
recall = recall_score(y_test, prediction)
print("Precision:", precision)
print("Recall:", recall)
```

```python
from sklearn.metrics import f1_score

# F1-Score hesapla
f1 = f1_score(y_test, prediction)
print("F1-Score:", f1)
```
